```python
from pathlib import Path

# Define the path
train_path = Path(
    r"C:\Users\yousi\.cache\kagglehub\datasets\youssefsalahzakria\fruit-and-vegetables-classification\versions\5\train"
)

if train_path.exists():
    # Extract only the names of the directories
    folders = [folder.name for folder in train_path.iterdir() if folder.is_dir()]

    print(f"Detected folders: {folders}")
else:
    print("Path not found.")
```

```python
# List of folders identified as culinary fruits
# List of folders identified as culinary fruits
culinary_fruits = [
    "Apple", "Avocado", "Banana", "Blackberry", "Blueberry",
    "Dates", "Dragonfruit", "Fig", "Grapes", "Guava",
    "Kiwi", "Lemon", "Mango", "Orange", "Pear",
    "Pineapple", "Pomegranate", "Rambutan", "Strawberry", "Watermelon"
]

print(f"{'Fruit Name':<20} | {'Files':<8} | {'Size (MB)':<10}")
print("-" * 45)

total_overall_size = 0

for fruit in culinary_fruits:
    fruit_path = train_path / fruit

    if fruit_path.exists() and fruit_path.is_dir():
        # Get list of files
        files = [f for f in fruit_path.iterdir() if f.is_file()]
        file_count = len(files)

        # Calculate total size in bytes, then convert to MB
        # 1 MB = 1024 * 1024 bytes
        folder_size_bytes = sum(f.stat().st_size for f in files)
        folder_size_mb = folder_size_bytes / (1024 * 1024)

        total_overall_size += folder_size_bytes

        print(f"{fruit:<20} | {file_count:<8} | {folder_size_mb:<10.2f}")
    else:
        print(f"{fruit:<20} | Path not found")

print("-" * 45)
print(f"{'Total Weight':<31} | {total_overall_size / (1024 * 1024):.2f} MB")
```

```python
import shutil
import os
from pathlib import Path

# 1. Setup Paths
source_root = Path(r"C:\Users\yousi\.cache\kagglehub\datasets\youssefsalahzakria\fruit-and-vegetables-classification\versions\5")
destination_root = Path("data")

# 2. List of Culinary Fruits to extract
culinary_fruits = [
    "Apple", "Avocado", "Banana", "Blackberry", "Blueberry",
    "Dates", "Dragonfruit", "Fig", "Grapes", "Guava",
    "Kiwi", "Lemon", "Mango", "Orange", "Pear",
    "Pineapple", "Pomegranate", "Rambutan", "Strawberry", "Watermelon"
]

# 3. Subfolders to process
subsets = ["train", "test", "validation"]

print("Starting extraction...")

for subset in subsets:
    # Create the destination subset folder (e.g., data/train)
    target_subset_path = destination_root / subset
    target_subset_path.mkdir(parents=True, exist_ok=True)

    source_subset_path = source_root / subset

    if not source_subset_path.exists():
        print(f"Skipping {subset}: Source folder not found.")
        continue

    for fruit in culinary_fruits:
        source_fruit_dir = source_subset_path / fruit
        target_fruit_dir = target_subset_path / fruit

        if source_fruit_dir.exists():
            # Check if it's already been copied to avoid errors
            if not target_fruit_dir.exists():
                # Use shutil.copytree to copy or shutil.move to move
                shutil.copytree(source_fruit_dir, target_fruit_dir)
                print(f"[{subset}] Copied: {fruit}")
            else:
                print(f"[{subset}] Already exists: {fruit}")
        else:
            print(f"[{subset}] Warning: {fruit} not found in source.")

print("\nFinished! Your 'data' folder is ready.")
```
